#### DATA (example per category)

```python
age_dist = pd.read_csv("CBS - bevolking; geslacht, leeftijd, postcode 2016-2018.csv", sep = ";", index_col = "ID", dtype = str)
imm_dist = pd.read_csv("CBS - geslacht, migratieachtergrond, postcode 2014-2016.csv", sep = ";", index_col = "ID", dtype = str)
pos_house = pd.read_csv("CBS - geslacht, positie huishouden, postcode 2014-2019.csv", sep = ";", index_col = "ID", dtype = str)
household = pd.read_csv("CBS - huishoudenssamenstelling, postcode 1998-2019.csv", sep = ";", index_col = "ID", dtype = str)
```

Cleaning age_dist

```python
age_dist_repl = {"Geslacht" : {"T001038" : "Total", "3000" : "Men", "4000" : "Women"}, 
                 "Leeftijd" : {"10000" : "Total", "22000" : "95<", "70100" : "0-5", "70200" : "5-10", 
                               "70300" : "10-15", "70400" : "15-20", "70500" : "20-25", "70600" : "25-30",
                               "70700" : "30-35", "70800" : "35-40", "70900" : "40-45", "71000" : "45-50", 
                               "71100" : "50-55", "71200" : "55-60", "71300" : "60-65", "71400" : "65-70",
                               "71500" : "70-75", "71600" : "75-80", "71700" : "80-85", "71800" : "85-90",
                               "71900" : "90-95"},
                 "Postcode" : {"NL01" : "NL_total", "PC0999" : "Unknown"}}

for col in age_dist.columns:
    age_dist[col] = age_dist[col].str.strip() 

age_dist.replace(age_dist_repl, inplace = True)
age_dist["Postcode"] = age_dist["Postcode"].str.replace("PC", "")
age_dist["Perioden"] = age_dist["Perioden"].str.replace("JJ00", "").astype(int)
age_dist["Bevolking_1"] = age_dist["Bevolking_1"].astype(float)
```

Cleaning pos_house

```python
pos_house_repl = {"Geslacht" : {"T001038" : "Total", "3000" : "Men", "4000" : "Women"},
                  "PositieInHetHuishouden" : {"1015100" : "Single", "1015110" : "Single_parent", 
                                              "1015120" : "Couple_kids", "1015130" : "Couple_no_kids",
                                              "1016440" : "Child_at_home", "1016510" : "Other", 
                                              "1050001" : "Person_private_household",
                                              "1050002" : "Person_institutional_household", "T009002" : "Total"}, 
                  "Postcode" : {"NL01" : "NL_total", "PC0999" : "Unknown"}}

for col in pos_house.columns:
    pos_house[col] = pos_house[col].str.strip()
    
pos_house.replace(pos_house_repl, inplace = True)
pos_house["Postcode"] = pos_house["Postcode"].str.replace("PC", "")
pos_house["Perioden"] = pos_house["Perioden"].str.replace("JJ00", "").astype(int)
pos_house["Bevolking_1"] = pos_house["Bevolking_1"].astype(float)
```

Clean household

```python
household_repl = {"Huishoudenssamenstelling" : {"1016030" : "Multiple_w_children", "1016040" : "Multiple_no_children",
                                               "1050010"  : "Total_private_household", "1050015" : "Single_person"}, 
                  "Postcode" : {"NL01" : "NL_total", "PC0999" : "Unknown"},
                 "GemiddeldeHuishoudensgrootte_2" : {"." : np.nan}}

for col in household.columns:
    household[col] = household[col].str.strip()
    
household.replace(household_repl, inplace = True)
household["Postcode"] = household["Postcode"].str.replace("PC", "")
household["Perioden"] = household["Perioden"].str.replace("JJ00", "").astype(int)
household[["ParticuliereHuishoudens_1", "GemiddeldeHuishoudensgrootte_2"]] = household[["ParticuliereHuishoudens_1", "GemiddeldeHuishoudensgrootte_2"]].astype(float)
```
